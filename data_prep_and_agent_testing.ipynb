### Subtask:
Analyze the 'product_reviews' data to extract sentiment and identify key aspects mentioned in reviews for each product. This information can be used to enrich product representations or as a separate tool for sentiment analysis.


**Reasoning**:
Select the 'Review' and 'Product_Standardized' columns from the `final_merged_df` and initialize a sentiment analysis model.

The previous command failed because `nltk.downloader.DownloadError` does not exist, and the 'vader_lexicon' was not found. The traceback suggests using `nltk.download('vader_lexicon')`. I will fix the error handling and ensure the lexicon is downloaded before proceeding with sentiment analysis.



```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

# Download VADER lexicon if not already downloaded
try:
    # Check if the resource is available
    nltk.data.find('sentiment/vader_lexicon.zip')
except LookupError:
    # If not found, download it
    print("VADER lexicon not found. Downloading...")
    nltk.download('vader_lexicon')
    print("Download complete.")

# Select relevant columns
reviews_df = final_merged_df[['Review', 'Product_Standardized']].copy()

# Drop rows where 'Review' is NaN, as these cannot be analyzed for sentiment
reviews_df.dropna(subset=['Review'], inplace=True)

# Initialize VADER sentiment intensity analyzer
analyzer = SentimentIntensityAnalyzer()

print("Reviews DataFrame with selected columns:")
display(reviews_df.head())

# Apply sentiment analysis and store results
sentiment_results = []
for index, row in reviews_df.iterrows():
    review = row['Review']
    product = row['Product_Standardized']
    sentiment_scores = analyzer.polarity_scores(str(review)) # Ensure review is string
    sentiment_results.append({
        'Product_Standardized': product,
        'Review_Sentiment_Compound': sentiment_scores['compound'],
        'Review_Sentiment_Positive': sentiment_scores['pos'],
        'Review_Sentiment_Neutral': sentiment_scores['neu'],
        'Review_Sentiment_Negative': sentiment_scores['neg']
    })

sentiment_df = pd.DataFrame(sentiment_results)

print("\nSentiment Analysis Results per Review:")
display(sentiment_df.head())

# Group by product and calculate aggregated sentiment metrics
aggregated_sentiment_df = sentiment_df.groupby('Product_Standardized').agg(
    average_sentiment_compound=('Review_Sentiment_Compound', 'mean'),
    average_sentiment_positive=('Review_Sentiment_Positive', 'mean'),
    average_sentiment_neutral=('Review_Sentiment_Neutral', 'mean'),
    average_sentiment_negative=('Review_Sentiment_Negative', 'mean'),
    review_count=('Review_Sentiment_Compound', 'count')
).reset_index()

print("\nAggregated Sentiment Analysis Results per Product:")
display(aggregated_sentiment_df.head())
```

```python
!pip install -q -U google-generativeai langchain langgraph langchain-google-genai

import os
import getpass
from typing import List, Dict, Any, TypedDict

from langchain_core.runnables import RunnableParallel, RunnablePassthrough
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.graph import StateGraph, END
import pandas as pd # Ensure pandas is imported

# Helper function to set up Google API Key
# def setup_google_api():
#     if "GOOGLE_API_KEY" not in os.environ:
#         os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Google API Key: ")

# # Call the setup function
# setup_google_api()
os.environ["GOOGLE_API_KEY"] = 'AIzaSyDlo-rq9YxnUs4r5sJEHlgMw1dgs5TAd08'
index = faiss.read_index('/content/product_embeddings.faiss')

# Define the State
class PersonalShopperState(TypedDict):
    input: str  # User's original query
    input_type: str | None  # 'keyword', 'vague', or 'good'
    clarification_questions: List[str] | None # Questions to ask the user
    clarification_answers: Dict[str, str] | None # User's answers to questions (used if user provides answers later)
    recommended_products: List[Dict[str, Any]] | None # List of recommended products
    follow_up_question: str | None # Question to ask after recommendations for keyword inputs

# Define the LLM
llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash-latest", temperature=0)

# Define the Nodes

def check_input_type(state: PersonalShopperState) -> PersonalShopperState:
    """
    Checks the input type ('keyword', 'vague', or 'good') using an LLM.
    - 'keyword': Specific product type (e.g., serum, cleanser). Requires recommendation AND follow-up question.
    - 'vague': General query needing clarification (e.g., gentle for summer). Requires clarification questions ONLY.
    - 'good': Specific enough to recommend, but not just a single product type (e.g., moisturizer for dry skin). Requires recommendation ONLY.
    The LLM should output ONLY one of these three words.
    """
    print(f"Checking input type for: {state['input']} using LLM")
    input_text = state['input']

    # Prompt for the LLM to classify the input
    prompt = f"""You are a helpful assistant that classifies user skincare queries into one of three types: 'keyword', 'vague', or 'good'.

    Here are the definitions:
    - 'keyword': The user is asking for a specific product category, like "serum", "cleanser", or "moisturizer".
    - 'vague': The user's query is general and needs more information to provide a good recommendation, like "something gentle for summer" or "products for glowing skin".
    - 'good': The user's query is specific enough to recommend products without asking for more information right away, like "anti-aging serum for oily skin" or "moisturizer for dry skin".

    Analyze the following user query and output ONLY one word: 'keyword', 'vague', or 'good'.

    User Query: "{input_text}"

    Classification:"""

    try:
        response = llm.invoke(prompt)
        # Attempt to parse the LLM's response, expecting one of the three words
        classified_type = response.content.strip().lower()

        if classified_type in ['keyword', 'vague', 'good']:
            input_type = classified_type
            print(f"LLM classified input as: {input_type}")
        else:
            # Fallback if LLM output is unexpected
            print(f"LLM returned unexpected output: '{classified_type}'. Defaulting to 'good'.")
            input_type = 'good' # Defaulting to 'good' on unexpected output

    except Exception as e:
        print(f"An error occurred during LLM classification: {e}. Defaulting to 'good'.")
        input_type = 'good' # Defaulting to 'good' on error


    new_state = state.copy()
    new_state['input_type'] = input_type
    return new_state

def ask_clarification_questions(state: PersonalShopperState) -> PersonalShopperState:
    """
    Generates clarification questions based on a vague input.
    """
    print("Generating clarification questions for vague input...")
    input_text = state['input']

    # Use the LLM to generate clarification questions based on the brand persona and vague input
    prompt = f"""You are a helpful and friendly salesperson for EverGlow Labs. The customer is asking about "{input_text}". This is a bit vague. To help them find the perfect product, generate 1-2 concise clarification questions. Focus on understanding their skin type, specific concerns, or preferred product types (like serum, toner, SPF). Present them clearly.

    Clarification Questions:
    """
    response = llm.invoke(prompt)
    # Assuming the LLM returns questions separated by newlines or similar
    questions = [q.strip() for q in response.content.strip().split('\n') if q.strip()]
    print(f"Generated questions: {questions}")

    new_state = state.copy()
    new_state['clarification_questions'] = questions
    # We set recommended_products to None here because for vague inputs, we only ask questions
    new_state['recommended_products'] = None
    new_state['follow_up_question'] = None
    return new_state


# Keep receive_clarification defined, although it's not directly linked in the graph
# for the 'vague' flow's immediate execution path as requested ("ask only").
# It would be used if the user provides answers in a *subsequent* step.
def receive_clarification(state: PersonalShopperState) -> PersonalShopperState:
    """
    Simulates receiving the user's answers to clarification questions.
    In a real Next.js integration, this would involve receiving user input
    from the frontend based on the questions previously asked.
    This function definition is kept but not part of the direct 'vague' path ENDing.
    """
    print("Simulating receiving clarification answers...")
    # This node would typically be triggered by an external event (user input)
    # and the state would be updated with the answers before invoking the graph again.
    # For this example, we just acknowledge it.
    print("Clarification answers would be processed here if provided by the user in a follow-up interaction.")

    # In a real scenario, you'd update the state like this:
    # new_state = state.copy()
    # new_state['clarification_answers'] = received_answers # Assume received_answers is passed or retrieved
    # return new_state

    # For this example, we just return the current state as it's not directly reachable
    # in the defined graph flow after 'ask_clarification_questions'.
    return state


# Placeholder for the function to search the FAISS index and get product details
# This function needs access to comprehensive_product_profiles_df, index, and embedding_model globals.
def search_products_with_index(query: str, top_k: int = 5) -> List[Dict[str, Any]]: # Increased top_k for better margin selection chance
    """
    Searches the FAISS index with the query and returns product details.
    Assumes access to comprehensive_product_profiles_df, index, and embedding_model globals.
    """
    print(f"Searching FAISS index for query: '{query}'")
    global comprehensive_product_profiles_df # Assuming it's a global variable
    global index
    global embedding_model # Assuming it's a global variable

    if comprehensive_product_profiles_df is None or index is None or embedding_model is None:
        print("Error: FAISS index, embedding model, or product data not available for search.")
        return []

    try:
        # Generate embedding for the query
        query_embedding = embedding_model.encode(query).astype('float32')

        # Perform search in FAISS index
        # Reshape query_embedding for faiss search (needs to be 2D)
        distances, indices = index.search(query_embedding.reshape(1, -1), top_k)

        # Get the indices of the top_k results
        result_indices = indices[0]

        # Retrieve the full product details for these indices
        # Ensure indices are within the bounds of the dataframe
        valid_indices = [idx for idx in result_indices if idx < len(comprehensive_product_profiles_df)]
        if not valid_indices:
             print("No valid indices found in FAISS search.")
             return []

        # Get the product details from the comprehensive DataFrame using the indices
        # Use iloc to get rows by integer position
        search_results_df = comprehensive_product_profiles_df.iloc[valid_indices].copy()

        # Convert to a list of dictionaries
        search_results = search_results_df.to_dict('records')

        print(f"Found {len(search_results)} initial potential products from search.")
        return search_results

    except Exception as e:
        print(f"An error occurred during FAISS search: {e}")
        return []


def recommend_products(state: PersonalShopperState) -> PersonalShopperState:
    """
    Recommends 1-3 products using the FAISS index based on the input
    and any clarification answers, and prioritizes by margin.
    Uses the comprehensive_product_profiles_df and the FAISS index.
    """
    print("Recommending products using FAISS index...")
    input_text = state['input']
    clarification_answers = state.get('clarification_answers', {})
    recommended_products = []

    global comprehensive_product_profiles_df # Assuming it's a global variable
    global index # Assuming it's a global variable
    global embedding_model # Assuming it's a global variable
    global llm # Assuming it's a global variable

    if comprehensive_product_profiles_df is None or index is None or embedding_model is None or llm is None:
         print("Error: Necessary data or models not available for recommendation.")
         new_state = state.copy()
         new_state['recommended_products'] = [] # Return empty list of recommendations
         return new_state

    # Combine input text and clarification answers for the recommendation query
    combined_query = input_text
    if clarification_answers:
        # Include clarification answers in the query for search
        combined_query += " " + " ".join(clarification_answers.values())
        print(f"Using combined query for FAISS search: {combined_query}")
    else:
         print(f"Using original input for FAISS search: {combined_query}")

    # --- Product Search using FAISS Index ---
    # Use the search_products_with_index function to get initial candidates.
    # We search for more than 3 to allow for margin-based sorting.
    initial_search_results = search_products_with_index(combined_query, top_k=10)

    if initial_search_results:
        # Convert the search results back to a DataFrame for easier sorting and selection
        search_results_df = pd.DataFrame(initial_search_results)

        # Ensure 'margin' column exists and is numeric before sorting
        if 'margin' in search_results_df.columns:
             # Convert margin to numeric, coercing errors to NaN and dropping rows where margin is not a valid number
            search_results_df['margin'] = pd.to_numeric(search_results_df['margin'], errors='coerce')
            search_results_df.dropna(subset=['margin'], inplace=True)

            if not search_results_df.empty:
                # Sort by margin (descending)
                search_results_df.sort_values(by='margin', ascending=False, inplace=True)

                # Select the top 1-3 products after sorting
                recommended_products = search_results_df.head(3).to_dict('records')
                print(f"Found and recommending {len(recommended_products)} products based on FAISS search and margin.")
            else:
                print("Warning: No products left after dropping rows with invalid margin values.")
                # If no products have valid numeric margin, just take the top 3 from original results (if any)
                recommended_products = initial_search_results[:3]
                print(f"Recommending {len(recommended_products)} products based on FAISS similarity only.")


        else:
            print("Warning: 'margin' column not found in search results. Recommending based on FAISS similarity only (top 3).")
            # If margin is not available, just take the top 3 from the initial search
            recommended_products = initial_search_results[:3]
            print(f"Recommending {len(recommended_products)} products based on FAISS similarity only.")


        # Add a short justification for each recommended product if any were found
        if recommended_products:
            print("Generating justifications...")
            for product in recommended_products:
                # Generate a short justification using LLM
                justification_prompt = f"""As a helpful and friendly salesperson for EverGlow Labs, write a concise justification (maximum 15 words) for recommending the product '{product.get('name', 'N/A')}' based on its key details and how it might address a customer's needs.
Brand Philosophy: "EverGlow Labs exists to prove that nature and science can co-author skincare that actually works. We formulate every product around three uncompromising pillars: 1. Plant-Powered, Clinically Proven... 2. Radical Transparency... 3. Barrier-First, Planet-First..."
Product Details: Name: {product.get('name', 'N/A')}, Category: {product.get('category', 'N/A')}, Key Ingredients: {product.get('top_ingredients', 'N/A')}, Description snippet: {product.get('description', 'N/A')[:100]}...
Customer Query: {state['input']}
Clarification Info: {state.get('clarification_answers', 'None provided')}

Justification (max 15 words):"""
                justification_response = llm.invoke(justification_prompt)
                # Truncate to roughly 15 words
                justification_text = justification_response.content.strip()
                product['justification'] = ' '.join(justification_text.split()[:15]) + ('...' if len(justification_text.split()) > 15 else '')

                print(f"Justification for {product.get('name', 'N/A')}: {product['justification']}") # For demonstration
        else:
             print("No products to generate justifications for.")

    else:
        print("No products found matching the criteria using FAISS index.")
        recommended_products = []


    new_state = state.copy()
    new_state['recommended_products'] = recommended_products
    # Ensure clarification questions are None after recommendation, unless they were asked and answered
    # (which isn't the flow for vague inputs anymore)
    if state.get('input_type') != 'vague': # Don't overwrite if clarification was asked for vague input in a previous step
         new_state['clarification_questions'] = None

    return new_state

def generate_follow_up_question(state: PersonalShopperState) -> PersonalShopperState:
    """
    Generates a follow-up question only if the original input_type was 'keyword'.
    """
    print("Checking if follow-up question is needed...")
    input_type = state['input_type'] # Use the classified input type

    if input_type != 'keyword':
        print("Follow-up question not needed for this input type.")
        new_state = state.copy()
        new_state['follow_up_question'] = None # Ensure follow-up is None if not keyword
        return new_state

    print("Generating follow-up question with salesperson persona...")
    input_text = state['input']
    recommended_products_names = ", ".join([p.get('name', 'a product') for p in state.get('recommended_products', [])])


    # Prompt to generate a follow-up question for keyword input after recommendations
    prompt = f"""You are a helpful and friendly salesperson for EverGlow Labs. The customer initially searched for "{input_text}" and you've just shown them recommendations like {recommended_products_names}. To make sure I'm helping them find the absolute best product for their unique skin, what is one more quick question I could ask about their skin type or specific goals? Keep it friendly and concise (1-2 sentences).

    Follow-up Question:
    """

    response = llm.invoke(prompt)
    follow_up_question = response.content.strip()

    new_state = state.copy()
    new_state['follow_up_question'] = follow_up_question
    print(f"Generated follow-up question: {follow_up_question}")

    return new_state


# Define the Graph
workflow = StateGraph(PersonalShopperState)

# Add nodes
workflow.add_node("check_input_type", check_input_type)
workflow.add_node("ask_clarification_questions", ask_clarification_questions)
# receive_clarification node definition is kept, but not directly in this flow's graph structure
# workflow.add_node("receive_clarification", receive_clarification)
workflow.add_node("recommend_products", recommend_products)
workflow.add_node("generate_follow_up_question", generate_follow_up_question)

# Set entry point
workflow.set_entry_point("check_input_type")

# Add edges
# From check_input_type:
# If vague -> ask_clarification_questions
# If keyword or good -> recommend_products
workflow.add_conditional_edges(
    "check_input_type",
    lambda state: {
        "vague": "ask_clarification_questions",
        "keyword": "recommend_products",
        "good": "recommend_products"
    }.get(state.get('input_type'), "recommend_products") # Default to recommend if classification fails
)

# From ask_clarification_questions:
# Always go to END for the 'vague' case as requested ("ask 1-2 follow up questions only")
workflow.add_edge("ask_clarification_questions", END)

# From recommend_products:
# If original input_type was keyword -> generate_follow_up_question
# If original input_type was good -> END
workflow.add_conditional_edges(
    "recommend_products",
    lambda state: "generate_follow_up_question" if state.get('input_type') == 'keyword' else END
)

# From generate_follow_up_question:
# Always go to END
workflow.add_edge("generate_follow_up_question", END)

# Compile the graph
app = workflow.compile()


# --- Integration with Next.js (Conceptual) ---
# 1. Initial Query: Frontend sends user query to backend. Backend initializes state, calls `app.invoke(initial_state)`.
# 2. Vague Query -> Clarification:
#    - If `check_input_type` determines 'vague', it transitions to `ask_clarification_questions`.
#    - `ask_clarification_questions` generates questions, updates state.
#    - The workflow transitions to `END`.
#    - Your backend needs to detect the final state contains `clarification_questions` and send them back to the frontend.
#    - Frontend displays questions to the user.
#    - User answers in frontend. Frontend sends answers back to backend.
#    - Backend creates a *new* initial state for the *next* interaction, including the original query and the received answers in `clarification_answers`. It would then call `app.invoke(new_initial_state)`. The graph would likely need a different entry point or logic to handle starting with answers (e.g., going directly to `recommend_products`).
# 3. Keyword/Good -> Recommendation:
#    - If `check_input_type` determines 'keyword' or 'good', it transitions directly to `recommend_products`.
#    - `recommend_products` finds and ranks products, updates state with `recommended_products`.
#    - If input was 'keyword', it transitions to `generate_follow_up_question`.
#    - If input was 'good', it transitions to `END`.
# 4. Follow-up Question (for Keyword):
#    - `generate_follow_up_question` creates the question, updates state.
#    - Transitions to `END`.
# 5. Final Response:
#    - When the workflow reaches `END`, your backend extracts `recommended_products` and `follow_up_question` (if any) from the final state.
#    - Backend sends this information back to the frontend to display to the user.

# Example of invoking the graph:

# Example 1: Vague Query (will ask clarification questions and END)
print("--- Example 1: Vague Query ---")
initial_state_vague = {"input": "something gentle for summer", "input_type": None, "clarification_questions": None, "clarification_answers": None, "recommended_products": None, "follow_up_question": None}
# This invocation will generate questions and immediately end the workflow.
final_state_vague = app.invoke(initial_state_vague)

print("\nFinal State (Vague Query):")
# display(final_state_vague)

print("\n" + "="*50 + "\n") # Separator

# Example 2: Keyword Query (will recommend and ask follow-up)
print("--- Example 2: Keyword Query ---")
initial_state_keyword = {"input": "serum", "input_type": None, "clarification_questions": None, "clarification_answers": None, "recommended_products": None, "follow_up_question": None}
final_state_keyword = app.invoke(initial_state_keyword)

print("\nFinal State (Keyword Query):")
# display(final_state_keyword)

print("\n" + "="*50 + "\n") # Separator

# Example 3: Good Query (will recommend products only)
print("--- Example 3: Good Query ---")
initial_state_good = {"input": "moisturizer for dry skin", "input_type": None, "clarification_questions": None, "clarification_answers": None, "recommended_products": None, "follow_up_question": None}
final_state_good = app.invoke(initial_state_good)

print("\nFinal State (Good Query):")
# display(final_state_good)

print("\n" + "="*50 + "\n") # Separator

# Example 4: Another Good Query
print("--- Example 4: Another Good Query ---")
initial_state_good_2 = {"input": "anti-aging cream with retinol", "input_type": None, "clarification_questions": None, "clarification_answers": None, "recommended_products": None, "follow_up_question": None}
final_state_good_2 = app.invoke(initial_state_good_2)

print("\nFinal State (Good Query 2):")
# display(final_state_good_2)
```
